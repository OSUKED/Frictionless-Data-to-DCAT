```python
import pandas as pd

from jinja2 import Template
import re
import requests

import matplotlib.pyplot as plt

from rdflib import Graph, URIRef
import networkx as nx
```

```python
template_fp = 'templates/tabular-schema.rdf'

with open(template_fp, 'r') as template_file:
    template = Template(template_file.read())

template
```

```python
# example_datapackage_url = 'https://github.com/AyrtonB/EveryNoise-Watch/raw/main/data/datapackage.json'
example_datapackage_url = 'https://data.open-power-system-data.org/conventional_power_plants/2020-10-01/datapackage.json'
    
datapackage = requests.get(example_datapackage_url).json()

len(datapackage['resources'])
```

```python
populated_template = template.render(datapackage=datapackage)
populated_fp = 'populated/tabular-schema.rdf'

with open(populated_fp, 'w') as populated_file:
    populated_template = re.sub('(?<=\n)\s+(?=\n)', '', populated_template).replace('\n\n', '\n')
    populated_file.write(populated_template)
```

```python
g = Graph()

g.parse(populated_fp, publicID='_')

g
```

```python
datasets = []

for node in list(g.all_nodes()):
    if isinstance(node, URIRef) == True:
        if node[:10] == '_:dataset-':
            datasets += [node]
            
datasets
```

```python
triplets = []

for subject, predicate, object_ in g:
    triplets += [{'subject':subject, 'predicate':predicate, 'object':object_}]
    
df_triplets = pd.DataFrame(triplets)

df_triplets#.head()
```

```python
from rdflib.extras.external_graph_libs import rdflib_to_networkx_multidigraph

G = rdflib_to_networkx_multidigraph(g)

# Plot Networkx instance of RDF Graph
fig, ax = plt.subplots(dpi=150)

pos = nx.spring_layout(G, scale=10, k=10, iterations=500)
edge_labels = nx.get_edge_attributes(G, 'r')
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels, ax=ax)
nx.draw(G, with_labels=True, ax=ax)
```
