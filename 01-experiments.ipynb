# Table Schema

The Tabular Data Resource provides:
* The name property to identify the dataset.
* The path property to define the relative path to the CSV file.
* The schema property following the Table Schema.
* The profile property with the value tabular-data-resource.
* The dialect property with the CSV Dialect.
* Optionally, the encoding property when the value is not UTF-8.
* Optionally, the title, description, hash, licences, etc.
* Optionally, the format (i.e.) "csv" and media type "text/csv".

The schema property in turn provides:
* The fields property with at least one field descriptor.
* Optionally, missing value rules, primary key and foreign keys.

And each field provides:
* The name property with a unique identifier matching the CSV column.
* Optionally, type, format, title and description.
* Optionally, the constraints property to aid a validator.

N.b. these notes were taken from [this post](https://www.seachess.net/notes/dive-into-frictionlessdata)

### Graph Utilities

```python
#exports
def add_predicates_and_objects(
    g,
    root_node,
    dict_,
    predicates_to_objects: dict={}
):
    """Adds the dataset contributors to the catalog graph
    """ 
    
    for predicate, (object_type, object_loc) in predicates_to_objects.items():
        if object_loc in dict_.keys():
            g.add((root_node, predicate, object_type(dict_[object_loc])))
            
    return g
```

### Table Schema Graph Generation

We'll begin by initialising the graph with a single `dcat:Catalog` node

```python
#exports
def initialise_catalog(g: rdflib.graph.Graph=Graph()) -> rdflib.graph.Graph:
    """Initialises the graph and adds the catalog root node
    """
    if g is None:
        g = Graph()
        
    catalog_node = BNode(value='catalog')
    g.add((catalog_node, RDF.type, DCAT.Catalog))
    
    return g
```

```python
g = initialise_catalog()

g
```

<br>

Next we'll add the high-level root attributes

```python
def add_root_attrs(
    g: rdflib.graph.Graph=Graph(), 
    datapackage: dict={},
    predicates_to_objects: dict={
         DCAT['name']: (Literal, 'name'),
         DCAT['title']: (Literal, 'title'),
         DCAT['description']: (Literal, 'description'),
         PROV['qualifiedAttribution']: (Literal, 'attribution'),
         DCAT['issued']: (Literal, 'publicationDate'),
         DCAT['identifier']: (Literal, 'id'),
         FOAF['homepage']: (Literal, 'homepage'),
         DCAT['modified']: (Literal, 'lastChanges'),
         DCAT['landingPage']: (Literal, 'documentation'),
     }
) -> rdflib.graph.Graph:   
    """Adds the dataset root attributes to the catalog graph
    """ 
    
    catalog_node = rdflib.term.BNode('catalog')
    
    if (catalog_node, None, None) not in g:
        g = initialise_catalog(g)
        
    root_attrs = {k: v for k, v in datapackage.items() if not isinstance(v, (dict, list))}        
    g = add_predicates_and_objects(g, catalog_node, root_attrs, predicates_to_objects)
                
    return g
```

```python
g = add_root_attrs(g, datapackage)
```

<br>

Then the contributors

```python
def add_contributors(
    g: rdflib.graph.Graph=Graph(), 
    datapackage: dict={},
    predicates_to_objects: dict={
         FOAF.name: (Literal, 'title'),
         FOAF.mbox: (URIRef, 'email'),
         FOAF.workInfoHomepage: (URIRef, 'path'),
     }
) -> rdflib.graph.Graph:   
    """Adds the dataset contributors to the catalog graph
    """ 
    
    catalog_node = rdflib.term.BNode('catalog')
    
    if (catalog_node, None, None) not in g:
        g = initialise_catalog(g)
        
    for i, contributor in enumerate(datapackage['contributors']):
        contributor_id = f'contributor-{i}'
        contributor_node = BNode(value=contributor_id)

        g.add((contributor_node, RDF.type, FOAF.Person))
        g.add((catalog_node, DC.contributor, contributor_node))
        
        g = add_predicates_and_objects(g, contributor_node, contributor, predicates_to_objects)
                
    return g
```

```python
g = add_contributors(g, datapackage)
```

<br>

The resources

```python
#exports
def add_resource_fields(g, dataset_node, dataset, dataset_id, field_predicates_to_objects, field_types_to_RDF_URI):
    """Adds the dataset resource fields to the catalog graph
    """ 
    
    if 'schema' in dataset.keys():
        if 'fields' in dataset['schema'].keys():
            for i, column in enumerate(dataset['schema']['fields']):
                column_id = f'{dataset_id}--column-{i}'
                column_node = BNode(value=column_id)

                g.add((column_node, RDF.type, CSVW.column))
                g.add((column_node, CSVW.table, dataset_node))

                g = add_predicates_and_objects(g, column_node, column, field_predicates_to_objects)

                column_type = column['type']
                if column_type in field_types_to_RDF_URI.keys():
                    g.add((column_node, CSVW.table, field_types_to_RDF_URI[column_type]))
                else:
                    warn(f'The field type `{column_type}` was not recognised')
                    
    return g

def add_resources(
    g: rdflib.graph.Graph=Graph(), 
    datapackage: dict={},
    pathed_predicates_to_objects: dict={
        'root': {
            DCAT['title']: (Literal, 'name'),
            DCAT['description']: (Literal, 'title'),
            DCAT['byteSize']: (Literal, 'bytes'),
            DCAT['downloadURL']: (URIRef, 'path'),
            DCAT['format']: (Literal, 'format'),
            DCAT['MediaType']: (Literal, 'mediatype'),
            CSVW['encoding']: (Literal, 'encoding'),
        },
        'dialect': {
            CSVW['delimiter']: (Literal, 'delimiter'),
            CSVW['lineTerminators']: (Literal, 'lineTerminator'),
        },
        'schema': {
            CSVW['primaryKey']: (Literal, 'primaryKey'),
            CSVW['foreignkeys']: (Literal, 'foreignKey'),
        }
    },
    field_predicates_to_objects: dict={
        DCAT['name']: (Literal, 'name'),
        DCAT['title']: (Literal, 'description'),
    },
    field_types_to_RDF_URI: dict={
        'string': XSD.string,
        'number': XSD.float,
        'boolean': XSD.boolean,
        'integer': XSD.integer,
    },
    profiles_to_schemas: dict={
        'tabular-data-resource': 'https://specs.frictionlessdata.io/schemas/table-schema.json'
    }
) -> rdflib.graph.Graph:
    """Adds the dataset resources to the catalog graph
    """ 
    
    # checking prescence of `catalog_node`
    catalog_node = rdflib.term.BNode('catalog')
    
    if (catalog_node, None, None) not in g:
        g = initialise_catalog(g)
    
    # creating metadata for each dataset
    for i, dataset in enumerate(datapackage['resources']):
        # preparing dataset and distribution nodes
        dataset_id = f'dataset-{i}'
        distribution_id = f'distribution-{i}'
        
        dataset_node = BNode(value=dataset_id)
        distribution_node = BNode(value=distribution_id)
        
        g.add((catalog_node, DCAT.dataset, dataset_node))
        g.add((dataset_node, RDF.type, DCAT.Dataset))
        g.add((dataset_node, DCAT.distribution, distribution_node))
        g.add((distribution_node, RDF.type, DCAT.Distribution))

        # simple attribute setting        
        for path, predicates_to_objects in pathed_predicates_to_objects.items():
            if path == 'root':
                g = add_predicates_and_objects(g, distribution_node, dataset, predicates_to_objects)
                
            else:
                if path in dataset.keys():
                    g = add_predicates_and_objects(g, distribution_node, dataset[path], predicates_to_objects)
        
        # field attribute setting
        g = add_resource_fields(g, distribution_node, dataset, dataset_id, field_predicates_to_objects, field_types_to_RDF_URI)                                        

        # complex attribute setting
        if 'profile' in dataset.keys():
            schema = profiles_to_schemas[dataset['profile']]
            g.add((distribution_node, DCAT['schema'], URIRef(schema)))
            
        if 'dialect' in dataset.keys():
            if 'header' in dataset['dialect'].keys():
                header_row_count = int(bool(dataset['dialect']['header'].capitalize()))
                g.add((distribution_node, CSVW['headerRowCount'], Literal(header_row_count)))
                
    return g
```

```python
g = add_resources(g, datapackage)
```

<br>

And the original data sources

```python
#exports
def add_sources(
    g: rdflib.graph.Graph=Graph(), 
    datapackage: dict={},
    predicates_to_objects: dict={
         DCAT.title: (Literal, 'title'),
         DCAT.downloadURL: (URIRef, 'path'),
     }
) -> rdflib.graph.Graph:
    """Adds the dataset sources to the catalog graph
    """ 
    
    catalog_node = rdflib.term.BNode('catalog')
    
    if (catalog_node, None, None) not in g:
        g = initialise_catalog(g)
    
    for i, source in enumerate(datapackage['sources']):
        source_id = f'source-{i}'
        source_node = BNode(value=source_id)
        
        g.add((catalog_node, PROV.hadPrimarySource, source_node))
        g.add((source_node, RDF.type, DCAT.Distribution))
        
        g = add_predicates_and_objects(g, source_node, source, predicates_to_objects)
                
    return g
```

```python
g = add_sources(g, datapackage)
```

<br>

We'll then combine these steps within the instantiation of a `tableSchema` Class

```python
#exports
class tableSchema(Graph):
    def __init__(self, datapackage, g=None):
        super().__init__()
        self = initialise_catalog(self)

        self = add_root_attrs(self, datapackage)
        self = add_contributors(self, datapackage)
        self = add_resources(self, datapackage)
        self = add_sources(self, datapackage)
        
    def to_frame(self):
        triplets = []

        for subject, predicate, object_ in g:
            triplets += [{'subject':subject, 'predicate':predicate, 'object':object_}]

        return pd.DataFrame(triplets)
```

```python
g = tableSchema(datapackage)
```

We can also use `networkx` and `matplotlib` to visualise the graph, though it is a little crowded...

```python
G = rdflib_to_networkx_multidigraph(g)

# Plot Networkx instance of RDF Graph
fig, ax = plt.subplots(dpi=150)

pos = nx.spring_layout(G, scale=10, k=10, iterations=500)
edge_labels = nx.get_edge_attributes(G, 'r')
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels, ax=ax)
nx.draw(G, with_labels=True, ax=ax)
```
